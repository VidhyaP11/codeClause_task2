# SELF DRIVING CAR

```python
import pygame
#intializing
pygame.init()
#For displaying window
win=pygame.display.set_mode((1200, 400))
#For loading track and car image
trk=pygame.image.load('track.png')
car=pygame.image.load('car.png')
#change the size of car
car=pygame.transform.scale(car, (30,60))
#decalring the start position of car
c_X=155 #150 can use
c_Y=300
# offset of a car
off_x=0
off_y=0
#for camera focusing
foc_dis=25
dri=True
direction='up'
#For slowing the car speed
clock=pygame.time.Clock()
while dri:
    #for fixing the stop postion
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            dri=False
    
    clock.tick(60)
    #camera setting
    cam_x=c_X+off_x+15
    cam_y=c_Y+off_y+15
    #for direction moving
    u_p=win.get_at((cam_x, cam_y-foc_dis))[0]
    R_p=win.get_at((cam_x+foc_dis,cam_y))[0]
    d_p=win.get_at((cam_x,cam_y+foc_dis))[0]
    print(u_p, R_p, d_p)
    
    #changing direction
    if direction =='up' and u_p!=255 and R_p ==255:
        direction ='right'
        #this 30 for camera position 60-30
        off_x=30
        car = pygame.transform.rotate(car, -90)
    elif direction =='right' and R_p!=255 and d_p ==255:
        direction ='down'
        #this is for changing the car position because it is oustside track in x direction
        c_X=c_X+30
        #this is for changing camera postion in this the camera outside the track in x direction
        off_x=0
        off_y=30
        car = pygame.transform.rotate(car, -90)
    elif direction =='down' and d_p!=255 and R_p ==255:
        direction ='right'
        #this is for changing the car position because it is oustside track in y direction 
        c_Y=c_Y+30
        #this is for changing camera postion in this the camera outside the track in y direction
        off_y=0
        off_x=30
        car = pygame.transform.rotate(car, 90)
    elif direction =='right' and R_p!=255 and u_p ==255:
        direction ='up'
        #this is for changing the car position because it is oustside track in y direction 
        c_X=c_X+30
        #changing the camera position
        off_x=0
        car = pygame.transform.rotate(car, 90)
    
    if direction =='up'and u_p==255:
        c_Y=c_Y-2
    elif direction == 'right' and R_p==255:
        c_X=c_X+2
    elif direction == 'down' and d_p==255:
        c_Y=c_Y+2
    win.blit(trk, (0,0))
    win.blit(car,(c_X,c_Y))
    #for creating camera
    pygame.draw.circle(win, (0 ,255 ,0),(cam_x,cam_y),5,5)
    pygame.display.update()
```
